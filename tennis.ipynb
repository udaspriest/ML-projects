```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load and investigate the data here:
path = 'tennis_stats.csv'
df = pd.read_csv(path)
#print(df.head())
service_game_columns = ['Aces','DoubleFaults','Firstserve','FirstServePointsWon','SecondServePointsWon','BreakPointsFaced','BreakPointsSaved','ServiceGamesPlayed','ServiceGamesWon','TotalServicePointsWon']
return_game_columns = ['FirstServeReturnPointsWon','SecondServeReturnPointsWon','BreakPointsOpportunities','BreakPointsConverted','ReturnGamesPlayed','ReturnGamesWon','TotalPointsWon']
outcomes = ['Wins','losses','Winnings','Ranking']
```

```python
# exploratory analysis:
plt.scatter(df[service_game_columns[7]],df[outcomes[0]],alpha=0.4)
plt.show()
```

```python
## single feature linear regression:
lm = LinearRegression()
x= df[['ServiceGamesPlayed']]
y = df[['Wins']]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
lm.fit(x_train,y_train)
print(lm.score(x_test,y_test))
y_predict = lm.predict(x_test)
plt.scatter(x_test,y_test)
plt.plot(x_test,y_predict)
plt.show()
```

```python
## two feature linear regression:
lm2 = LinearRegression()
x = df[['BreakPointsOpportunities','ReturnGamesPlayed']]
y = df['Wins']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
lm2.fit(x_train,y_train)
print(lm2.score(x_test,y_test))
```

```python
## multiple feature linear regression:
mlr = LinearRegression()
x=df[['Aces','DoubleFaults','FirstServePointsWon','SecondServePointsWon','BreakPointsFaced','BreakPointsSaved','ServiceGamesPlayed','ServiceGamesWon','TotalServicePointsWon','FirstServeReturnPointsWon','SecondServeReturnPointsWon','ReturnGamesWon','TotalPointsWon']]
y=df[['Winnings']]
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.3)
mlr.fit(x_train,y_train)
print(mlr.score(x_test,y_test))
```
